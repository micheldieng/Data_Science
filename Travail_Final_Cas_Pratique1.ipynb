# 10. Dans cette partie, on demandera d’écrire un programme qui permet de :

## • créer et remplir un vecteur représentant la somme pluviométrique annuelle
## • déterminer la moyenne du vecteur obtenu
## • calculer la différence entre le vecteur et sa moyenne

```python
pl = dt.iloc[:, 2:]
pl
```

```python
annual_pluvio = np.sum(pl, axis = 1)
ecart_type = annual_pluvio - np.nanmean(annual_pluvio)
print("La somme pluviométrique annuelle est de :\n", annual_pluvio)
print("\n Sa moyenne est : ", np.nanmean(annual_pluvio))
print("\nL'ecart-type de la pluviometrie est de : \n",ecart_type )
```

# 11. Tracer la courbe de la différence obtenue à la question précédente en fonction des années

```python
fig4 ,ax4 = plt.subplots(figsize = (15,8))
plt.bar(dt.Annees, ecart_type, label = "Ecart-type annual pluvio")
plt.title("Difference entre la pluvio annuelle et sa moyenne ")
plt.legend()
plt.grid()
plt.show()
```
