# 2. En se basant sur les tableaux des coefficients, écrire un programme qui cacule la moyenne de chaque UE et détermine si l’étudiant l’a validée ou pas.

```python
Phy=deliberation.iloc[:,2]*(2.5/6)+deliberation.iloc[:,3]*(2.5/6)+deliberation.iloc[:,4]*(1/6)
Math=deliberation.iloc[:,5]*(2.5/5)+deliberation.iloc[:,6]*(2.5/5)
Chim=deliberation.iloc[:,7]*(2.5/6)+deliberation.iloc[:,8]*(2.5/6)+deliberation.iloc[:,9]*(1/6)
Info=deliberation.iloc[:,10]*(2.5/5)+deliberation.iloc[:,11]*(2.5/5)
Base=deliberation.iloc[:,12]*(1.5/4)+deliberation.iloc[:,13]*(1.5/4)+deliberation.iloc[:,14]*(1/4)
Gra=deliberation.iloc[:,15]*(2/4)+deliberation.iloc[:,16]*(2/4)
```

```python
P=[]
for i in range(len(Phy)):
    if Phy[i]<10:
        P.append("NV")
    else:
        P.append("V")
M=[]
for i in range(len(Math)):
    if Math[i]<10:
        M.append("NV")
    else:
        M.append("V")
C=[]
for i in range(len(Chim)):
    if Chim[i]<10:
        C.append("NV")
    else:
        C.append("V")
I=[]
for i in range(len(Info)):
    if Info[i]<10:
        I.append("NV")
    else:
        I.append("V")
B=[]
for i in range(len(Base)):
    if Base[i]<10:
        B.append("NV")
    else:
        B.append("V")
G=[]
for i in range(len(Gra)):
    if Gra[i]<10:
        G.append("NV")
    else:
        G.append("V")
```

# 3. Écrire un programme qui insère une colonne de validation (V : UE validée, NV : UE nonvalidée) après chaque UE.

```python
deliberation1=deliberation.iloc[:,0:2]
deliberation1.insert(2,"Physique",Phy,allow_duplicates=False)
deliberation1.insert(3,"DecisionPhysique",P,allow_duplicates=False)
deliberation1.insert(4,"Maths",Math,allow_duplicates=False)
deliberation1.insert(5,"DecisionMaths",M,allow_duplicates=False)
deliberation1.insert(6,"Chimie",Chim,allow_duplicates=False)
deliberation1.insert(7,"DecisionChimie",C,allow_duplicates=False)
deliberation1.insert(8,"Info",Info,allow_duplicates=False)
deliberation1.insert(9,"DecisionInfo",I,allow_duplicates=False)
deliberation1.insert(10,"Base",Base,allow_duplicates=False)
deliberation1.insert(11,"DecisionBase",B,allow_duplicates=False)
deliberation1.insert(12,"Graphique",Gra,allow_duplicates=False)
deliberation1.insert(13,"DecisionGraphique",G,allow_duplicates=False)
deliberation1
```
