```python
import pandas as pd
fichier = "Del_Annuel.xlsx"
xl = pd.ExcelFile(fichier)
etud1 = xl.parse("PV")
etud1
```

```python
df0=etud1.dropna(how="any")
df0
```

```python
col=etud1.iloc[9,:]
col
```

```python
df=df0.rename(columns=col)
df
```

```python
df.dropna(inplace=True)
df.columns=col
df.columns=df.columns.fillna("Decision")
df.rename(columns={'RESULTAT': 'Total_Credit'},inplace=True)
df
```

```python
df.reset_index(drop=True)
```

```python
df_tri=df.sort_values(by=["Total_Credit","Nom","Prénom"],ascending=[False,True,True])
df_tri
```

```python
import pandas as pd
writer = pd.ExcelWriter("deliberation_ESTI_UAM.xlsx", engine="xlsxwriter")
df_tri.to_excel(writer, "L1")
df_tri.to_excel(writer, "C")

writer.save()
```

```python
A=list(df_tri.Decision.value_counts())
A
```

```python
import numpy as np
a1=np.around((A[0]*100)/len(df),1)
print("Le pourcentage de passage defintif est",a1)
a2=np.around((A[1]*100)/len(df),1)+0.1
print("Le pourcentage de passage conditionnel est",a2)
a3=np.around((A[2]*100)/len(df),1)
print("Le pourcentage de passage redouble est",a3)
a4=np.around((A[3]*100)/len(df),1)
print("Le pourcentage de passage exclu est",a4)
a5=a1+a2+a3+a4
print("L'effectif total est de:",a5)
```

```python
import numpy as np
import pandas as pd
ar = np.array([[A[0], a1], [A[1], a2], [A[2],a3],[A[3],a4],[len(df),a5]])
df1 = pd.DataFrame(ar, index = ['Passage_Definitif', 'Passage_Conditionnel', 'Redouble', 'Exclu','Effectif_total'], columns = ['Effectifs', 'Pourcentage'])
df1
```

```python
import matplotlib.pyplot as plt
labels = 'Passage_Definitif', 'Passage_Conditionnel', 'Redouble', 'Exclu'
serie = [a1,a2,a3,a4]
separation = (0, 0, 0, 0) # Séparation des tranches
fig, ax = plt.subplots()
ax.pie(serie, explode=separation, labels=labels, autopct='%1.0f%%',
shadow=True, startangle=90)
ax.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
